# Functions

steps:
    1) function declaration
    2) function body
    3) function calling

```python
# Default function
def func():
    print('Hello World')
func()
```

```python
# Sequential argument/ positional argument
def func(name, fname):
    print(name)
    print(fname)
func('nabeel', 'muhammad iqbal')
```

```python
def func(num1, num2):
    print(num1 + num2)
func(5, 6)
```

```python
# Keyword argument
def bio(name, fname):
    print(name)
    print(fname)
bio(fname = 'muhammad saleem', name = 'sameer')
```

```python
def tax(price, tax = 0.02):
    tax_rate = price * tax
    print(price + tax_rate)
tax(100, 0.05)
```

```python
def bio(greeting, name):
    print(greeting, name)
bio('Hello', name = 'nabeel')
```

```python
dic = {
    1: {
        'name': 'Nabeel',
        'fname': "Muhammad Iqbal",
    },
    2: {
        'name': 'Sameer',
        'fname': "Muhammad Saleem",
    }
}
def func(dic_name, key, target_key):
    print(dic[key][target_key])
func(dic, 2, 'fname')
```

```python
def func(*c):
    print(sum(c))
func(4, 5, 6)
```

```python
def func(name, fname, **others):
    print(name)
    print(fname)
    for key, value in others.items():
        print(key, ":", value)
func('Nabeel', 'Muhammad Iqbal', age = 20)
```

```python
def func(num1, num2):
    return num1 + num2
func(5, 6)
```

```python
def func(num1, num2):
    return num1 + num2
add = func(1, 2)
print(add)
```

```python
def func(name, fname):
    print(name)
    print(fname)
func(name = "sameer", fname='Muhammad Saleem')
print(name)
```
